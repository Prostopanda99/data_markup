```python
#импортируем необходимые библеотеки 
import pandas as pd
import requests
```

```python
#создаем поиска 
def get_result_org(o_title, m_title, ):
    keys = {'regkey': 'd37e5f9c2df49556a580b1c3dc8dcc7a',
        'org_title':o_title,
        'mitypeTitle':m_title,
        'export_type': '2'}
    r = requests.get('http://731163-cj72200.tmweb.ru/vri/', params=keys)
    df = pd.DataFrame(r.json())
    return df
```

```python
#Получаем данные по  цаги 
df_tsagi=get_result_org('ЦАГИ','манометр')
display(df_tsagi)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings('ignore')

# --- подготовка данных ---

# Источник (МАИ / ЦАГИ)
source_counts = pd.Series({'МАИ (расходомеры)': len(df_mai), 'ЦАГИ (манометры)': len(df_tsagi)})

# Тип поверки
vri_counts = df_full['vriType'].fillna('Не указан').value_counts()

# Топ-15 шифров знака (поверочные организации)
cipher_counts = df_full['mi_signCipher'].fillna('Не указан').value_counts().head(15)

# Тип результирующего документа
doc_counts = df_full['result_doc_type'].fillna('Не указан').value_counts()

# --- построение ---
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Анализ поверок средств измерения (МАИ + ЦАГИ)', fontsize=16, fontweight='bold', y=1.01)

# 1. Соотношение источников (pie)
ax = axes[0, 0]
ax.pie(source_counts, labels=source_counts.index, autopct='%1.1f%%',
       colors=['#4C72B0', '#DD8452'], startangle=90,
       wedgeprops={'edgecolor': 'white', 'linewidth': 2})
ax.set_title('Соотношение МАИ / ЦАГИ', fontsize=13, fontweight='bold')

# 2. Тип поверки (horizontal bar)
ax = axes[0, 1]
bars = ax.barh(vri_counts.index, vri_counts.values, color='#4C72B0', edgecolor='white')
ax.set_title('Распределение по типу поверки', fontsize=13, fontweight='bold')
ax.set_xlabel('Количество записей')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', ' ')))
for bar, val in zip(bars, vri_counts.values):
    ax.text(bar.get_width() + max(vri_counts.values) * 0.01, bar.get_y() + bar.get_height() / 2,
            f'{val:,}'.replace(',', ' '), va='center', fontsize=9)
ax.margins(x=0.15)

# 3. Топ-15 шифров знака поверителя
ax = axes[1, 0]
bars = ax.barh(cipher_counts.index[::-1], cipher_counts.values[::-1], color='#DD8452', edgecolor='white')
ax.set_title('Топ-15 шифров поверителя (mi_signCipher)', fontsize=13, fontweight='bold')
ax.set_xlabel('Количество записей')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', ' ')))
for bar, val in zip(bars, cipher_counts.values[::-1]):
    ax.text(bar.get_width() + max(cipher_counts.values) * 0.01, bar.get_y() + bar.get_height() / 2,
            f'{val:,}'.replace(',', ' '), va='center', fontsize=9)
ax.margins(x=0.2)

# 4. Тип результирующего документа
ax = axes[1, 1]
bars = ax.barh(doc_counts.index, doc_counts.values, color='#55A868', edgecolor='white')
ax.set_title('Тип результирующего документа', fontsize=13, fontweight='bold')
ax.set_xlabel('Количество записей')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', ' ')))

# Перенос длинных меток
import textwrap
ax.set_yticklabels([textwrap.fill(lbl.get_text(), 28) for lbl in ax.get_yticklabels()])
for bar, val in zip(bars, doc_counts.values):
    ax.text(bar.get_width() + max(doc_counts.values) * 0.01, bar.get_y() + bar.get_height() / 2,
            f'{val:,}'.replace(',', ' '), va='center', fontsize=9)
ax.margins(x=0.2)

plt.tight_layout()
plt.savefig('vri_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"\nВсего записей в df_full: {len(df_full):,}".replace(',', ' '))
print(f"Колонки ({len(df_full.columns)}): {list(df_full.columns)}")
```
